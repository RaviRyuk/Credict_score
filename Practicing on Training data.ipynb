## Pickling of Preprocessing

```python
def dataclean1(df):
    df.drop(['x242','x295','x304','x098','x155','x259','x255','x257','x256'],axis=1,inplace=True)
    df.drop(['x094','x067','x096','x095'],axis=1,inplace=True)
    null_clms=df.loc[:,df.isna().any()]
    df=df.fillna(null_clms.median())
    return df
joblib.dump(dataclean1,'data.pkl')
```

## Train_Test and Validation set 

```python
df2=df.loc[0:69999,:]#train_test_split
df3=df.loc[70000:79999,:]#validationset
```

```python
print(df2.shape)
print(df3.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df2.drop('y',axis=1)
y=df2['y']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scx=StandardScaler()
x_train=scx.fit_transform(x_train)
x_test=scx.transform(x_test)
```

## Pickling of scaling

```python
joblib.dump(scx,'scale.pkl')
```

# Linear Regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lrg=LinearRegression()
```

```python
lrg.fit(x_train,y_train)
```

```python
pre=lrg.predict(x_test)
```

## Pickling of LinearRegression 

```python
joblib.dump(lrg,'Linear.pkl')
```

```python
pre=pd.Series(pre)
pre.name='Predicted'
y_test.name='Actual'
y_test.reset_index(inplace=True,drop=True)
```

```python
pd.concat([y_test,pre],axis=1)
```

# Metrics

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,pre))
print(mean_absolute_error(y_test,pre))
print(mean_squared_error(y_test,pre))
print(np.sqrt(mean_squared_error(y_test,pre)))
```

# validation set

```python
x_val=df3.drop('y',axis=1)
y_val=df3.y
x_val=scx.transform(x_val)
```

```python
pre_v=lrg.predict(x_val)
```

```python
print(r2_score(y_val,pre_v))
print(mean_squared_error(y_val,pre_v))
print(mean_absolute_error(y_val,pre_v))
print(np.sqrt(mean_squared_error(y_val,pre_v)))
```

```python
list(zip(list(y_val),list(pre_v)))
```

## Lasso

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
alpha = [10,100,1000,10000,100000]
for i in alpha:
    las = Lasso(alpha=i,normalize=True)
    las.fit(x_train,y_train)
    pre = las.predict(x_test)
    print(len(las.coef_[las.coef_ > 0]))
    print(r2_score(y_test,pre))
    print(mean_absolute_error(y_test,pre))
    print(mean_squared_error(y_test,pre))
    print(np.sqrt(mean_squared_error(y_test,pre)))
```

# Ridge

```python
alpha = [10,100,1000,10000,100000]
for i in alpha:
    las = Ridge(alpha=i,normalize=True)
    las.fit(x_train,y_train)
    pre = las.predict(x_test)
    print(len(las.coef_[las.coef_>0]))
    print(r2_score(y_test,pre))
    print(mean_absolute_error(y_test,pre))
    print(mean_squared_error(y_test,pre))
    print(np.sqrt(mean_squared_error(y_test,pre)))
```
